```python
plot.figure(figsize=(16,8))
plot.bar('Sub-Category','Category', data=df)
plot.title('Category vs Sub Category')
plot.xlabel('Sub-Catgory')
plot.ylabel('Category')
plot.xticks(rotation=45)
plot.show()
```

```python

df.hist(bins=50 ,figsize=(20,15))
plot.show();
```

```python
plot.figure(figsize=(15,15))
sns.countplot(x=df['State'])
plot.xticks(rotation=90)
plot.title("STATE")
plot.show()
```

```python
sns.set(style="whitegrid")
plot.figure(2, figsize=(20,15))
sns.barplot(x='Sub-Category',y='Profit', data=df, palette='Spectral')
plot.suptitle('Pie Consumption Patterns in the United States', fontsize=16)
plot.show()
```

As we can see in the graphs that Sub-Categories like Machines, Bookcases, Tables are in loss. So we can stop selling thiese products or try to find ome other way to make profit in these products. 

We can also limit our product to the cities which ha high or medium purchases of Sub-Categories products to remove transport and rental losses.
